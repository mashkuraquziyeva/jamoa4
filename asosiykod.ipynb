```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
wur = pd.read_csv("wur.csv")
```

```python
wur
```

```python
rank = wur['milliy_daraja']
plt.bar(10, height=5)
plt.show()
```

```python
talim_sifati = wur['oqitish_sifati']
fakultet_sifati = wur['fakultet_sifati']
plt.title("O'qitish Sifati vs Fakultet Sifati")
plt.plot(talim_sifati, color="DarkBlue", label="Ta'lim sifati")
plt.plot(fakultet_sifati, color="DarkGreen", label="Fakultet sifati")
plt.legend()
plt.show()
```

```python
daraja = wur['milliy_daraja']
plt.hist(daraja, bins=8, linewidth=0.5, edgecolor="white")
plt.title("Milliy daraja")
plt.show()
```

```python
ish = wur['bitiruvchilarni_ishga_joylashishi']
plt.stem(ish, label="Ishga joylashish ko'rsatkichi")
plt.title("Bitiruvchilarning ishga joylashishi")
plt.legend()
plt.show()
```

```python
hisob = wur['hisob']
plt.plot(hisob, 'ro', label="Hisob")
plt.title("Yakuniy hisob bo'yicha")
plt.legend()
plt.show()
```

```python
nashriyot = wur['nashriyoti']
plt.boxplot(nashriyot, 5)
plt.title("Universitetlar nashriyoti")
plt.show()
```
